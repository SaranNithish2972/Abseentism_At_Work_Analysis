# Fitting Models:

# KNN Algorithm

K Nearest Neighbour - simplest supervised machine learning algorithms used for classification.

Euclidean Distance.


It makes highly accurate predictions and is Used to solve both classification and regression problems.

```python
#Split arrays or matrices into random train and test subsets.
#sklearn package is provided by scikit-learn
from sklearn.model_selection import train_test_split
```

```python
#Defining x and y values
x = df[['Age','Years of Experience','Service Time(8*30 =240)','Target given for each employee (10)','Target Achieved by each employee (10)','Absenteeismworkinhours','Performer (1,2,3)','Disciplinary failure','WorkSatisfactionscore']]
y = df['Absenteeism']
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size=0.20, random_state=1)
```

```python
#Classifier implementing the k-nearest neighbors vote.
from sklearn.neighbors import KNeighborsClassifier
#Choosing the value K = 1
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(train_x,train_y)
```

```python
#predict the values of any new data points
pred = knn.predict(test_x)
```

```python
#Classification report is used to measure the quality of predictions
#How many predictions are True and how many are False.
#Confusion matrix to evaluate the accuracy of a classification.
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(test_y,pred))
```

Accuracy = No of correct predictions / Total predictions <br>
Precision = No of correctly predicted positive values / No of Positive prediction <br>
If Precision increases Recall decreases

```python
print(confusion_matrix(test_y,pred))
```

```python
#Numpy is used for mathematical calc and handling arrays
import numpy as np
#It is the ratio of the number of error units of data to the total number of units.
# Values above 0.80 is an indication of a good classifier.
error_rate = []
for i in range(1,10):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(train_x,train_y)
    pred_i = knn.predict(test_x)
    error_rate.append(np.mean(pred_i != test_y))
```

```python
#plotting the error rates
plt.figure(figsize=(10,6))
plt.plot(range(1,10),error_rate,color='blue', linestyle='dashed', marker='o', markerfacecolor='red', markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
```

From the above plot we can infer that the values 45,46,47 has the less error rate. So we take the value as K.

```python
# First taking 
# When K=1
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(train_x,train_y)
pred = knn.predict(test_x)
print('WITH K=1')
print(classification_report(test_y,pred))
```

```python
print(confusion_matrix(test_y,pred))
```

```python
# When K=3
# Choosing K from the plot where the error rate is very low
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(train_x,train_y)
pred = knn.predict(test_x)
print('WITH K=46')
print(classification_report(test_y,pred))
```

Taking k as 46 with less error value gives us the High accuracy.

```python
print(confusion_matrix(test_y,pred))
```

# Decision Tree Classifier

Decision Tree Classifier is an supervised learning algorithm which creates the classification model by building a decision tree.<br>
It uses Entropy value to build tree.

```python
#sklearn package is provided by scikit-learn
#Classification report is used to measure the quality of predictions
#How many predictions are True and how many are False.
#Confusion matrix to evaluate the accuracy of a classification.
from sklearn.metrics import classification_report,confusion_matrix, accuracy_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
#Defining x and y values
x = df[['Age','Years of Experience','Service Time(8*30 =240)','Target given for each employee (10)','Target Achieved by each employee (10)','Absenteeismworkinhours','Performer (1,2,3)','Disciplinary failure','WorkSatisfactionscore']]
y = df['Absenteeism']
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size=0.20, random_state=1)
```

```python
#Choosing entropy for classification
#Gini index can also be choosen
df_model = DecisionTreeClassifier(criterion='entropy') 
df_model.fit(train_x, train_y)
predict_y = df_model.predict(test_x)
print(classification_report(test_y, predict_y))
```

```python
print("Accuracy:", accuracy_score(predict_y, test_y))
```

```python
print("Confusion matrix: \n", confusion_matrix(predict_y, test_y))
```

```python
#Importing tree from sklearn for classification
from sklearn import tree
#For Plots
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(200, 100))
tree.plot_tree(df_model, filled=True, rounded=True)
```

```python
print(tree.export_text(df_model))
```

# Random Forest Classifier

Random Forest Classifier is an Supervised Learning algorithm used to solve classification problems.<br>
Random Forest is made up of Many Decision Trees.

```python
#sklearn package is provided by scikit-learn
#Classification report is used to measure the quality of predictions
#How many predictions are True and how many are False.
#Confusion matrix to evaluate the accuracy of a classification.
from sklearn.metrics import classification_report,confusion_matrix, accuracy_score
```

```python
#Defining x and y values
x = df[['Age','Years of Experience','Service Time(8*30 =240)','Target given for each employee (10)','Target Achieved by each employee (10)','Absenteeismworkinhours','Performer (1,2,3)','Disciplinary failure','WorkSatisfactionscore']]
y = df['Absenteeism']
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size=0.20, random_state=1)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier(n_estimators=100)
rf_model.fit(train_x, train_y)
predict_y_3 = rf_model.predict(test_x)
print(classification_report(test_y, predict_y_3))
```

```python
print("Accuracy:", accuracy_score(predict_y_3, test_y))
print("Confusion matrix: \n", confusion_matrix(predict_y_3, test_y))
```
